```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize=(8.5, 6.5))

files = [("results.csv", r"N = $10^5$ points, local"),
         ("results_cluster.csv", r"N = $10^5$ points, cluster")] 

dfs = []
for file, info in files:
    try:
        dfs.append((pd.read_csv(file), info))
    except:
        pass

ax = fig.add_subplot(111)
ax.set_title(r'Acceleration depending on threads amount')
ax.set_xlabel('Threads, amount')
ax.set_ylabel('Acceleration')

for df, info in dfs:
    ax.plot(df.values[:, 0], df.values[0, 1]/df.values[:, 1], label=info)

ax.set_xticks(np.arange(1, 21))
ax.legend()
ax.grid()

plt.tight_layout()
plt.show()
```
